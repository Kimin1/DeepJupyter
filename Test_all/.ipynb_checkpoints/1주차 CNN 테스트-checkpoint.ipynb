## 3. 데이터 구조 변경하기

```python
import numpy as np
x_train = np.reshape(x_train, (50000,28,28,1))
x_val = np.reshape(x_val, (10000,28,28,1))
x_test = np.reshape(x_test, (10000,28,28,1))

print(x_train.shape)
print(x_test.shape)
```

## 4. 데이터 정규화



elongated contour: 길쭉한 윤곽
spherical contour: 구면 윤곽

```python
x_train = x_train.astype('float32')
x_val = x_val.astype('float32')
x_test = x_test.astype('float32')

gray_scale = 255
x_train /= gray_scale
x_val /= gray_scale
x_test /= gray_scale
```

## 4. 실제값을 원 핫 인코딩으로 변경하기



categorical: 범주형

```python
num_classes = 10
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_val = tf.keras.utils.to_categorical(y_val, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

## CNN 텐서플로우로 구현하기

```python
x = tf.placeholder(tf.float32, shape=[None, 28, 28, 1])
y_ = tf.placeholder(tf.float32, shape=[None, 10])
```

파라미터(weight, bias) 초기값 설정

## 난수 텐서



truncated: 잘린

왜 truncated를 쓰느냐?

표준 편차가 0.1인 값들 정규 분포 형태의 난수 텐서를 생성한다.
weight_cariable에서 shape은 conv1, conv2, fc1, fc2에서 원하는 가중치 변수값을
지정할 때 사용한다.



결국 필터를 만든다는 것은 가중치를 생성하는 것이다. 
shape은 필터의 기본적인 사이즈이고
weight는 변수이기 때문에 값이 학습할 때 마다 바뀌게 된다.

```python
def weight_variable(shape):
  initial = tf.truncated_normal(shape, stddev=0.1)
  return tf.Variable(initial)

def bias_variable(shape):
  initial = tf.constant(0.1, shape=shape)
  return tf.Variable(initial)
```

### 스트라이드, 패딩 설정



input으로 [배치, 높이, 넓이, 채널]을 가진 shape을 건네고, 필터로는 [필터 높이, 필터 넓이, 인 채널, 아웃채널]을 준다  
그리고 필터를 2차원의 모양으로 평면화 한다.   [필터 높이x필터 넓이x 채널, 아웃채널]  

그리고 입력 텐서에서 이미지 패치를 추출하여 모양의 가상 텐서를 형성합니다.

```python
def conv2d(x, W):
  return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')

def max_pool_2x2(x):
  return tf.nn.max_pool(x, ksize=[1, 2, 2, 1],
                        strides=[1, 2, 2, 1], padding='SAME')
```

conv1 레이어 설정

```python
W_conv1 = weight_variable([5, 5, 1, 16]) # 5x5x1 conv, 16 output
b_conv1 = bias_variable([16])
```

활성화 함수 설정

```python
h_conv1 = tf.nn.sigmoid(conv2d(x, W_conv1) + b_conv1)
```

풀링레이어 적용

```python
h_pool1 = max_pool_2x2(h_conv1)
```

activation map의 사이즈는 14 x 14가 되고 CON2 레이어 생성

```python
W_conv2 = weight_variable([5, 5, 16, 32])
b_conv2 = bias_variable([32])

h_conv2 = tf.nn.sigmoid(conv2d(h_pool1, W_conv2) + b_conv2)
h_pool2 = max_pool_2x2(h_conv2)
```

#### FC(Fully Connected Layer)

CONV를 통해 추출된 이미지의 특징들을 입력으로 받아 분류한다.

```python
W_fc1 = weight_variable([7 * 7 * 32, 128])
b_fc1 = bias_variable([128])

h_pool2_flat = tf.reshape(h_pool2, [-1, 7*7*32])
h_fc1 = tf.nn.sigmoid(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)
```

FC 영역에는 총 2개의 히든 레이어 존재, FC1은 128개, FC2는 10개의 노드

```python
W_fc2 = weight_variable([128, 10])
b_fc2 = bias_variable([10])

y_conv = tf.matmul(h_fc1, W_fc2) + b_fc2
```

크로스엔트로피를 설정

```python
cross_entropy = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits_v2(
        labels=y_, logits=y_conv))
```

### 옵티마이저를 사용해서 모델을 최적화

```python
train_step = tf.train.AdamOptimizer(0.001).minimize(cross_entropy)
```

### 정확도를 구하기 위해서 사용

- argmax  

tf.argmax(a, 1)는 2차원 배열의 각 행에서 가장 큰 값을 찾아  인덱스를 반환합니다.  

첫번째 행에서 10, 두번째 행에서 6, 세번째 행에서 8을 찾았기 때문에 실행결과가 [1 2 1]입니다.   

index               0   1  2   

      
               [[ 3  10  1]
                [ 4  5  6]
                [ 0  8  7]]

- reduce_mean



MNIST 코드에서는 정확도를 계산하기 위해서 tf.reduce_mean 함수가 사용됩니다.  

correct_prediction에 저장되어 있는 라벨값과 모델에서 계산된 값이 일치하는지 여부(True, False)를 tf.cast 함수를 사용하여 실수 1 또는 0으로 변환한 후, tf.reduce_mean를 사용하여 평균을 계산합니다.

- cast


```python
correct_prediction = tf.equal(tf.argmax(y_conv,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

#### 학습 및 테스트

미니배치를 사용하여 모델을 최적화. 미니배치는 총 500개의 학습데이터가 들어있고, 미니배치마다 파라미터를 조정


vectorization: 벡터화


#### 적당한 배치 사이즈를 줘야한다.

```python
# initialize
init = tf.global_variables_initializer()

# train hyperparameters
epoch_cnt = 3
batch_size = 500
iteration = len(x_train) // batch_size

# Start training
with tf.Session() as sess:
    tf.set_random_seed(777)
    # Run the initializer
    sess.run(init)
    for epoch in range(epoch_cnt):
        avg_loss = 0.
        start = 0; end = batch_size
        
        for i in range(iteration):
            if i%10 == 0:
                train_acc = accuracy.eval(
                        feed_dict={x:x_train[start: end], 
                                    y_: y_train[start: end]})
                print("step "+str(i)+ ": train acc: "+str(train_acc))
            train_step.run(
                feed_dict={
                    x:x_train[start: end], y_: y_train[start: end]})
            start += batch_size; end += batch_size    
        
        # Validate model
        val_accuracy = accuracy.eval(feed_dict={x:x_val, y_: y_val})
        print("validation accuracy: "+str(val_accuracy))
        
    test_accuracy = accuracy.eval(feed_dict={x:x_test, y_: y_test}) 
    print("test accuracy: "+str(test_accuracy))
```
